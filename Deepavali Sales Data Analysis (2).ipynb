```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:ax.bar_label(bars)
plt.title("Distribution of Genders")
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
def format_y_axis(value, tick_number):
    return f'{int(value):,}'
plt.figure(figsize=(16,8))
sales = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot(x='Gender',y='Amount',data=sales)
ax.yaxis.set_major_formatter(FuncFormatter(format_y_axis))
plt.title("Sales by Gender")
```

```python
plt.figure(figsize=(16, 8))
ax = sns.countplot(data=df,x='Age Group',hue = 'Gender')
for bars in ax.containers:ax.bar_label(bars)
plt.title("Age Group Distribution by Gender")
```

```python
plt.figure(figsize=(16 , 8))
sales = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot(x='Age Group',y='Amount',data=sales)
ax.yaxis.set_major_formatter(FuncFormatter(format_y_axis))

plt.title("Sales by Age Group")
```

```python
sns.set(rc={'figure.figsize':(23,10)})
ax = sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:ax.bar_label(bars)
plt.title("Distribution of Product_Category")
```

```python
sales = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax = sns.barplot(x='Product_Category',y='Amount' ,data=sales)
ax.yaxis.set_major_formatter(FuncFormatter(format_y_axis))
plt.title("Sales By Product_Category")
#for bars in ax.containers:ax.bar_label(bars)
```
